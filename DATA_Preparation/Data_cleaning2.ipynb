```python
import pandas as pd
import numpy as np
```

```python
%cd ~/Master-IV/PROYECTO_FM/
```

```python
aibb = pd.read_csv('airbnb1.csv')
aibb.shape
```

```python
# Selection of variables for Regression Model.
# New dataset
# We get rid of room type 'share room' to remove noise

aibb = aibb.drop(aibb[aibb.room_type=='Shared room'].index)
model = aibb[['availability_30','is_location_exact','zipcode','host_identity_verified',
              'bathrooms','bedrooms','beds','minimum_nights','maximum_nights','price','Price_Person_Night',
              'room_type','security_deposit','cleaning_fee','number_of_reviews','cancellation_policy','reviews_per_month','review_scores_rating',
               'review_scores_accuracy','review_scores_value','extra_people','guests_included','accommodates','host_is_superhost',
              'instant_bookable']]
```

```python
# CANCELATION POLICY
# Super_strict_30 and 60 are reclassified  as strict 
model.cancellation_policy.value_counts()

```

```python
model['cancellation_policy'] = model.cancellation_policy.astype(str)
model['cancellation_policy']= model.cancellation_policy.str.replace('super_strict_30','strict')
model['cancellation_policy']= model.cancellation_policy.str.replace('super_strict_60','strict')
```

```python
model.cancellation_policy.value_counts()
```

```python
# Zipcode Nas replaced by 0
model.zipcode.isnull().sum()
```

```python
model['zipcode'] = model.zipcode.replace(np.nan,0)
model.zipcode.value_counts()
```

```python
# Correction of wrong values
model['zipcode'] = model.zipcode.replace('2802\n28012',28012)
model['zipcode'] = model.zipcode.replace('28051\n28051',28051)
model['zipcode'] = model.zipcode.replace('28002\n28002',28002)
```

```python
# Replacement of security deposit NaNs
model['security_deposit'] = model.security_deposit.replace(np.nan,0)
model.security_deposit.isnull().sum()
```

```python
# Extra people conversion to EUR
model['extra_people'] = model.extra_people.str.replace('$','')
model['extra_people'] = model.extra_people.astype(float)
model['extra_people'] = model.extra_people/ 1.0651
```

```python
# remove space from room type name so that R can perform Linear regression
model['room_type'] =model.room_type.str.replace('Entire home/apt','Entire_home/apt')
model.room_type.value_counts()
```

```python
# Finally we replace NaNs for 0 

model = model.replace(np.nan,0)
model.isnull().sum()
```

```python
# File for Linear regression analysis in R
model.to_csv('model.csv')
```

```python
# Creation of dummy varibales for host_identity_verified
model_host = pd.get_dummies(model['host_identity_verified'])
m = model.join(model_host)
m.rename(columns={'f':'Host_identity_NOTverified','t':'host_identity_verified'},inplace=True)
m.drop(['host_identity_verified'],axis =1,inplace=True)
m.head()
```

```python
# Creation of dummy variables for room type 
room = pd.get_dummies(m['room_type'])
m = m.join(room)
m.drop(['room_type'],axis =1,inplace=True)
m.head()
```

```python
# One hot encoding
# Creation of dummy variables for cancellation policy
cancel = pd.get_dummies(model['cancellation_policy'])
m = m.join(cancel)
m.rename(columns={'flexible':'Cancel_policy_FLEXIBLE','moderate':'Cancel_policy_MODERATE','strict':'Cancel_policy_STRICT'},inplace=True)
m.drop(['cancellation_policy'],axis =1,inplace=True)
m.head()
```

```python
# Dummy for 'superhost'
super = pd.get_dummies(model['host_is_superhost'])
m = m.join(super)
m.rename(columns={'f':'NO_superhost','t':'YES_superhost'},inplace=True)
m.drop(['host_is_superhost'],axis =1,inplace=True)
m.head()
```

```python
# Dummy for 'instant bookable'
instant = pd.get_dummies(model['instant_bookable'])
m = m.join(instant)
m.rename(columns={'f':'NO_instant_book','t':'YES_instant_book'},inplace=True)
m.drop(['instant_bookable'],axis =1,inplace=True)
m.head()
```

```python
# Creation of dummy variables for 'location is exact'
loc = pd.get_dummies(m['is_location_exact'])
m = m.join(loc)
m.rename(columns={'f':'NO_location_exact','t':'YES_location_exact'},inplace=True)
m.drop(['is_location_exact'],axis =1,inplace=True)
m.head()
```
